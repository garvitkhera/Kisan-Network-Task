```python
from skimage import io, img_as_float64
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

# Brightness: 

```python
image = cv2.imread(r"C:\Users\GARVIT KHERA\Desktop\ImageProcessingChallenge\input\princeton_small.jpg")
im = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

# Edge Detection:

```python
# Using Canny function to do Canny Edge Detection

edges = cv2.Canny(image,100,200)
plt.subplot(221),plt.imshow(im, cmap = 'gray')
plt.title('Original Image')
plt.subplot(222),plt.imshow(edges, cmap = 'gray')
plt.title('Edge Image')
plt.show()
```

# Masking (Composite)

```python
# Read the images
foreground = cv2.imread(r'C:\Users\GARVIT KHERA\Desktop\ImageProcessingChallenge\input\comp_foreground.jpg')
background1 = cv2.imread(r'C:\Users\GARVIT KHERA\Desktop\ImageProcessingChallenge\input\comp_background.jpg')
alpha = cv2.imread(r'C:\Users\GARVIT KHERA\Desktop\ImageProcessingChallenge\input\comp_mask.jpg')
 
# Convert uint8 to float
foreground = foreground.astype(float)
background = background1.astype(float)
 
# Normalize the alpha mask to keep intensity between 0 and 1
alpha = alpha.astype(float)/255
 
# Multiply the foreground with the alpha matte
foreground = cv2.multiply(alpha, foreground)
 
# Multiply the background with ( 1 - alpha )
background = cv2.multiply(1.0 - alpha, background)
 
# Add the masked foreground and background.
outImage = cv2.add(foreground, background)
outImage = outImage/255

#Printing results
back = cv2.cvtColor(background1, cv2.COLOR_BGR2RGB)
back = img_as_float64(back)
plt.subplot(121)
plt.title('Original Image')
io.imshow(back)

skcomposite = img_as_float64(outImage)
plt.subplot(122)
plt.title('Composite Image')
io.imshow(skcomposite)
```

# Blurring an image:

```python
img = cv2.imread(r'C:\Users\GARVIT KHERA\Desktop\ImageProcessingChallenge\input\princeton_small.jpg') 
blurImg1 = cv2.blur(img,(2, 2))
blurImg1 = cv2.cvtColor(blurImg1, cv2.COLOR_BGR2RGB)
skimg1 = img_as_float64(blurImg1)
blurImg2 = cv2.blur(img,(5, 5))
blurImg2 = cv2.cvtColor(blurImg2, cv2.COLOR_BGR2RGB)
skimg2 = img_as_float64(blurImg2)
blurImg3 = cv2.blur(img,(10, 10))
blurImg3 = cv2.cvtColor(blurImg3, cv2.COLOR_BGR2RGB)
skimg3 = img_as_float64(blurImg3)
```

```python
plt.subplot(131)
plt.title('Blur Kernel 2')
io.imshow(skimg1)
plt.subplot(132)
plt.title('Blur Kernel 5')
io.imshow(skimg2)
plt.subplot(133)
plt.title('Blur Kernel 10')
io.imshow(skimg3)
io.show()
```

# Sharpening Image :

```python
im = cv2.imread(r'C:\Users\GARVIT KHERA\Desktop\ImageProcessingChallenge\input\princeton_small.jpg') 
kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
im = cv2.filter2D(im, -1, kernel)
img = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
img = img_as_float64(img)
plt.title('Sharpened Image')
io.imshow(img)
```
